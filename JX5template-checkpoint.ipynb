```python
def Firstquarter(conn):
    """
    List the quantity of the product in different month.
    """
    print("The Product in the first quarter:")
    with conn:
        sql1 = "SELECT Quantity_Ordered FROM Product_January WHERE product =?"
        sql2 = "SELECT Quantity_Ordered FROM Product_Feburary WHERE product =?"
        sql3 = "SELECT Quantity_Ordered FROM Product_March WHERE product =?"
        cur = conn.cursor()
        cur.execute(sql1,product)
        result_set1 = cur.fetchall()
        cur.execute(sql2,product)
        result_set2 = cur.fetchall()
        cur.execute(sql3,product)
        result_set3 = cur.fetchall()
        result_set=result_set1+result_set2+result_set3
        print("FirstQuantity", result_set)
```
